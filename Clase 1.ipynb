```python
import random #se importa la librera de python random

lametazo = ["Queridos", "Apreciados", "Distinguidos", "Honorables", "Estimados", "Respetados"]
potenciales_marranos = ["compatriotas", "conciudadanos", "copartidarios", "electores", "amigos"]
condicion = ["en mi gobierno", "con su apoyo", "siendo elegido", "si me siguen", "durante mi mandato"]
compromiso = ["voy a derrotar", "voy a derrotar","eliminar", "acabar", "luchar contra", "combatir"]
ilusion_Guerrerista = ["la violencia y ", "la delincuencia y", "la corrupcin y", "la inflacin y", "la pobreza y ", "el desplazamiento y"]
promesa =["trabajar por", "garantizar", "proteger", "velar por", "promover", "defender"]
beneficio_populista =["la educacin", "el empleo", "la paz", "la igualdad", "la salud"]
defendiendo_votos =["del pais", "de la ciudad", "de la comunidad", "de la poblacin", "para toda la gente", "de cada colombiano"]


lametazoo = random.choice(lametazo)
potenciales = random.choice(potenciales_marranos)
condi = random.choice(condicion)
compro = random.choice(compromiso)
ilusion = random.choice(ilusion_Guerrerista)
promesaa = random.choice(promesa)
beneficio = random.choice(beneficio_populista)
votos = random.choice(defendiendo_votos)

print("letra generada: " + lametazoo + " " + potenciales + " " + condi + " " + compro + " " + ilusion 
      + " " + promesaa + " " + beneficio+ " " + votos)
```
